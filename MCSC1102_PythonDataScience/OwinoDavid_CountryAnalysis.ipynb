```python
import pandas as pd
```

### load the dataset and display the first 5 and last 5 records.

```python
country_data=pd.read_csv('Country-data.csv')
country_data.head(5)
```

### Find and display the: Country with the highest GDP per capita (gdpp)

```python
country_data[['country', 'gdpp']].sort_values(by='gdpp', ascending=False).reset_index()['country'][0]
```

### Country with the lowest life expectancy (life_expec)

```python
country_data[['country', 'life_expec']].sort_values(by='life_expec', ascending=False).reset_index()['country'][0]
```

### Calculate the mean child mortality (child_mort) for countries with income greater than 10,000.

```python
countries=country_data[country_data['income']>10000]
countries['child_mort'].mean()
```

### Create a new column called health_investment_ratio computed as

Display the top 5 countries with the highest health_investment_ratio.

```python
country_data['highest health_investment_ratio']=(country_data['health']/country_data['income'])*100
country_data.head()
```

```python
country_data.head()
```

```python
### Display the top 5 countries with the highest health_investment_ratio.
```

```python
country_data[['country','highest health_investment_ratio']].sort_values(by='highest health_investment_ratio', ascending =False).head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Using Matplotlib, plot a bar chart showing the top 10 countries with the highest child mortality.

```python
child=country_data[['country','child_mort']].sort_values(by='child_mort', ascending=False).head(10).set_index('country')
child
```

```python
child.plot(kind='bar')
plt.show();
```

```python
### Plot a line graph of GDP per capita (gdpp) vs life expectancy (life_expec) for the top 15 economies by GDP.
```

```python
#Plot a line graph of GDP per capita (gdpp) vs life expectancy (life_expec) for the top 15 economies by GDP.
countries_gdp=country_data.sort_values(by='gdpp', ascending=False).head(15)
countries_gdp
```

```python
country_data['life_expec', 'gdpp']
```

```python
plt.plot(countries_gdp['life_expec'], countries_gdp['gdpp'])
plt.xlabel('life expectancy')
plt.ylabel('gdpp')
plt.show();
```

```python
Using Seaborn, create a correlation heatmap of the numerical columns. Highlight and comment on two interesting relationships you observe.

Answer text Question 1
```

```python
cols=['child_mort', 'exports', 'health', 'imports', 'income',
       'inflation', 'life_expec', 'total_fer', 'gdpp']
df_num=country_data[cols]
```

```python
# correlation matrix for numerical columns
corr_matrix = df_num.corr()

plt.figure(figsize=(12,6))
sns.heatmap(corr_matrix, annot=True, cmap='cividis', center=0)
plt.title('Heatmap')
```
