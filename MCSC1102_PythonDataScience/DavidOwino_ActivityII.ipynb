```python
# import library
import numpy as np
```

#### (a) Write a function that solves quadratic equations

```python
# create the function
def quadratic_eqn_solver (a, b, c):
    ''' 
    solve quadratic equations using quadratic formula
    does not solve quadratic equations with complex roots
    '''    
    if a == 0:
        return 'ZeroDivisionError: a cannot be zero hence not a quadratic equation'
    else:
        # makeing coefficient of x**2 1
        bb = b / a
        cc = c / a
        
        # create disciminant
        discriminant = (bb**2 - 4*cc)
        if discriminant < 0: 
            return 'Complex Roots'
        else:
            rdis = np.sqrt(discriminant)/2
            x1 = round((-bb/2 + rdis),3)
            x2 = round((-bb/2 - rdis),3)
            return x1, x2
    
# getting values from user
coef = [float(input(f'Enter the value of {i}:\n ')) for i in ['a', 'b', 'c']]

# Call solver
result = quadratic_eqn_solver(coef[0], coef[1], coef[2])

# Output result
if len(result) == 2:
    x1, x2 = result
    print(f'x = {x1}, x = {x2}')
else:
    print(result)
```

#### (b) Modify your function to handle the case of complex roots.

```python
# modify the function
def quadratic_eqn_solver (a, b, c):
    ''' 
    solve quadratic equation using quadratic formula
    also solve quadratic equations with complex roots
    '''    
    if a == 0:
        return 'ZeroDivisionError: a cannot be zero hence not a quadratic equation'
    else:
        # makeing coefficient of x**2 1
        bb = b / a
        cc = c / a
        
        # create disciminant
        discriminant = (bb**2 - 4*cc)
        if discriminant < 0: 
            mod_dis = -1*discriminant         
            rdis = round(np.sqrt(mod_dis) / 2, 3)
            real_part = round(-bb / 2, 3)
        
            # format the string for complex part 'i'
            x1 = f'{real_part} + {rdis}i'
            x2 = f'{real_part} - {rdis}i'
            return x1, x2
        else:
            rdis = np.sqrt(discriminant)/2
            x1 = round((-bb/2 + rdis),3)
            x2 = round((-bb/2 - rdis),3)
            return x1, x2
    
# getting values from user
coef = [float(input(f'Enter the value of {i}:\n ')) for i in ['a', 'b', 'c']]

# Call solver
result = quadratic_eqn_solver(coef[0], coef[1], coef[2])

# Output result
if len(result) == 2:
    x1, x2 = result
    print(f'x = {x1}, x = {x2}')
else:
    print(result)
```
