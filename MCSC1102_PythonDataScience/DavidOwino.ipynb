```python
sales = {

    'John': 120,

    'Mary': 95,

    'Peter': 75,

    'Lucy': 130,

    'Brian': 60,

    'Anne': 45,

    'James': 150,

    'Rita': 80

}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(sales, index=[1]).T
df_ok=df.reset_index().rename(columns={'index':'employee', 1:'sales'})
df_ok
```

```python
## Add code to compute and print the total and average sales for all employees. (5 Marks)
```

```python
df_ok['sales'].mean()
```

```python
 ## Identify and print the names of employees who met the target (sales ≥ 80) and those who did not meet the target (sales < 80). Use an if-else structure. (5 Marks)
```

```python
for i in df_ok['sales']:
    if i>=80:
        print(df_ok['employee'])
```

```python
if sales.values()<80:
    print(sales.keys())
```

```python
### Determine and print the employee(s) with the highest and lowest sales figures. (5 Marks) 
```

```python
#highest sales
df_ok[['employee', 'sales']].sort_values(by ='sales', ascending=False).reset_index()['employee'][0]
```

```python
# lowest sales
df_ok[['employee', 'sales']].sort_values(by ='sales').reset_index()['employee'][0]
```
