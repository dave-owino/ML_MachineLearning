Use the groupby() function to calculate the total sales made by each salesperson. Display the result in descending order of total sales. (6 Marks)


Group the data by Region and compute the average sales per region. (10 Marks)

```python
import pandas as pd
```

```python
data = {

    'Salesperson': ['Alice', 'Bob', 'Alice', 'Diana', 'Bob', 'Ethan', 'Alice', 'Diana', 'Ethan', 'Bob'],

    'Region': ['Nairobi', 'Mombasa', 'Kisumu', 'Nairobi', 'Nairobi', 'Kisumu', 'Mombasa', 'Kisumu', 'Nairobi', 'Mombasa'],

    'Sales': [25000, 18000, 22000, 27000, 15000, 20000, 30000, 23000, 24000, 16000]

}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df
```

#### Use the groupby() function to calculate the total sales made by each salesperson. Display the result in descending order of total sales. (6 Marks)

```python
data=df.groupby('Salesperson')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
data
```

#### Group the data by Region and compute the average sales per region. (10 Marks)

```python
df.groupby('Region')['Sales'].mean().reset_index()
```
