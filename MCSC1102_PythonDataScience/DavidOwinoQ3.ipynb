```python
# Branch C sales data

data_c = {

    'Branch': ['C', 'C', 'C'],

    'Product': ['Laptop', 'Keyboard', 'Headset'],

    'Sales': [140000, 8000, 9000]

}
```

```python
import pandas as pd
```

```python
data_a=pd.DataFrame(data_a)
data_a
```

```python
data_b=pd.DataFrame(data_b)
data_b
```

```python
data_c=pd.DataFrame(data_a)
data_c
```

```python
df=pd.concat([data_a, data_b, data_c])
df
```
